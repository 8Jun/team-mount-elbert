## Start of Data Cleaning.

### Removed all the special characters from the dataset 

```python

from pyspark.sql.functions import *
def cleanColumn(tmpdf,colName,findChar,replaceChar):
    tmpdf = tmpdf.withColumn(colName, regexp_replace(colName, findChar, replaceChar))
    return tmpdf

allColNames = news_df.schema.names
charToRemove= "[\"!@#$%^&*\(\)\{\}\[\]\'\'""',.?/:;-=+`~'...''..']"
replaceWith =""
for colName in allColNames:
    news_df=cleanColumn(news_df,colName,charToRemove,replaceWith)
```

```python
# Convert the words in lower case and remove stop words from stop_words
splitRDD_no_stop = splitRDD.filter(lambda x: x.lower() not in stop_words)

# Create a tuple of the word and 1 
splitRDD_no_stop_words = splitRDD_no_stop.map(lambda w: (w, 1))

# Count of the number of occurences of each word 
resultRDD = splitRDD_no_stop_words.reduceByKey(lambda x, y: x + y)
```

```python
# Swap the keys and values
resultRDD_swap = resultRDD.map(lambda x: (x[1], x[0]))

# Sort the keys in descending order
resultRDD_swap_sort = resultRDD_swap.sortByKey(ascending=False)

# Show the top 10 most frequent words and their frequencies
for word in resultRDD_swap_sort.take(10):
	print("{} has {} counts". format(word[1], word[0]))
```

```python
new_results_df = spark.createDataFrame(resultRDD_swap_sort)
new_results_df.createOrReplaceTempView("Word_Count")
```

```python
from textblob import TextBlob
import seaborn as sns
```

```python
pandas_df = news_df.toPandas()
pandas_df_noNA = pandas_df[pandas_df['authors'].notna()]
pandas_df_noNA
```

```python
#only include the top 4 authors from the data frame 'The Canadian Press','Cbc News','The Associated Press','Thomson Reuters'
top4_authors = pandas_df_noNA[pandas_df_noNA['authors'].isin(['The Canadian Press','Cbc News','The Associated Press','Thomson Reuters'])]
top4_authors
```

```python
#get polarity based on text column from the data frame
text_sample = top4_authors[['authors','text']].sample(1000)
def detect_polarity(text):
    return TextBlob(text).sentiment.polarity
text_sample['polarity'] = text_sample.text.apply(detect_polarity)
text_sample.head(100)
```

```python
#plot polarity in a histogram for all the authors
num_bins = 50
plt.figure(figsize=(10,6))
n, bins, patches = plt.hist(text_sample.polarity, num_bins, facecolor='blue', alpha=0.5)
plt.xlabel('Polarity', color='b', fontsize=20)
plt.ylabel('Count', color='b', fontsize=20)
plt.title('Histogram of Polarity for Covid-19 News Articles', color='b', fontsize=20)
plt.show();
```

```python
#box plot of polarity by authors
plt.figure(figsize=(10,6))
chart = sns.boxenplot(x='authors', y='polarity', data=text_sample)
plt.xlabel('Authors', color='c', fontsize=20)
plt.ylabel('Polarity', color='c', fontsize=20)
plt.title('Boxplot of Polarity for Covid-19 by Authors', color='c', fontsize=20)
plt.show(); 
```
